```python

import matplotlib.pyplot as plt
import numpy as np


imagen=plt.imread('Imagenes/redes.jpg')

plt.imshow(imagen)


print("Dimensiones de la Imagen: ")
print(imagen.shape)

print("\n")

#Al asignar cada valor a una variable, desempaquete la tupla de la siguiente manera.
alto, ancho, canales = imagen.shape
print('Ancho:  ', ancho)
print('Alto: ', alto)
print('Canales:', canales)


```

```python
plt.hist(imagen.ravel(), bins=256)
```

```python
#Declaramos la figura que contendra las diferentes imagenes de los canales
figura_canalesLuminosidad = plt.figure()

#-----------------IMAGEN CANAL ROJO DE LUMINOSIDAD---------------- 

#Agregamos la imagen a la posicion deseada de la figura con el .add_subplot
# .add_subplot(n°filas, n°columnas, n°grafica)
canalRojo_Luminosidad=figura_canalesLuminosidad.add_subplot(1,3,1)

#Seteamos el canal de luminosidad deseado
canalRojo_Luminosidad = imagen[:, :, 0]

#ploteamos nuestra imagen
plt.imshow(canalRojo_Luminosidad)

#Seteamos un tamaño mas grande por grafica
plt.rcParams["figure.figsize"] = (16, 16)

#Titulo
plt.title('Canal Rojo')

plt.hist(canalRojo_Luminosidad)



#-----------------IMAGEN CANAL VERDE DE LUMINOSIDAD---------------- 

canalVerde_Luminosidad=figura_canalesLuminosidad.add_subplot(1,3,2)

canalVerde_Luminosidad = imagen[:, :, 1]

plt.imshow(canalVerde_Luminosidad)

plt.rcParams["figure.figsize"] = (16, 16)

plt.title('Canal Verde')

#Ya que tenemos a modo de referencia los ejes de la primera imagen, vamos a sacarlos para
#el resto y ahorrar espacio
plt.axis("off")

plt.hist(canalVerde_Luminosidad)



#-----------------IMAGEN CANAL AZUL DE LUMINOSIDAD---------------- 

canalAzul_Luminosidad=figura_canalesLuminosidad.add_subplot(1,3,3)

canalAzul_Luminosidad = imagen[:, :, 2]

plt.imshow(canalAzul_Luminosidad)

plt.rcParams["figure.figsize"] = (16, 16)

plt.title('Canal Azul')

plt.axis("off")

plt.hist(canalAzul_Luminosidad)
```

```python
#-----------------IMAGEN CANAL AZUL DE LUMINOSIDAD---------------- 

canalAzul_Luminosidad=figura_canalesLuminosidad.add_subplot(1,1,1)

canalAzul_Luminosidad = imagen[:, :, 2]

plt.imshow(canalAzul_Luminosidad)

plt.rcParams["figure.figsize"] = (16, 16)

plt.title('Canal Azul')

plt.axis("off")

plt.hist(canalAzul_Luminosidad, orientation='horizontal')

```
