```python
'''
Pseudocolor puede ser una herramienta útil para mejorar el contraste
y visualizar sus datos más 
fácilmente. Esto es especialmente útil al realizar presentaciones de
sus datos utilizando proyectores
su contraste suele ser bastante pobre.
El pseudocolor solo es relevante para imágenes de luminosidad de 
un solo canal, escala de grises.
Actualmente tenemos una imagen RGB. Dado que R, G y B son similares 
(compruébelo usted mismo más 
arriba o en sus datos), podemos elegir un canal de nuestros datos
'''
import matplotlib.pyplot as plt
import numpy as np

imagen=plt.imread('Imagenes/fisica.jpg')

plt.imshow(imagen)

print("Dimensiones de la Imagen: ")
print(imagen.shape)
```

```python
#declaramos la figura que contendra las diferentes imagenes de los canales
figura_canalesLuminosidad=plt.figure()



#-----------------------------IMAGEN CANAL ROJO LUMINOSIDAD--------------------
#Agregamos la imagen a la posicion deseada de la figura con el .add_subplot
#.add_subplot(n°filas, n°columnas, n°grafica)
canalRojoLuminosidad=figura_canalesLuminosidad.add_subplot(1,3,1)

canalRojoLuminosidad=np.copy(imagen)

canalRojoLuminosidad=imagen[:,:,0]

#Ploteamos la imagen
plt.imshow(canalRojoLuminosidad)

#Seteamos un tamaño mas grande por grafica(lo declaramos una sola vez)
plt.rcParams["figure.figsize"]=(16,16)

plt.title("Canal Rojo")


#-----------------------------IMAGEN CANAL VERDE LUMINOSIDAD--------------------
#Agregamos la imagen a la posicion deseada de la figura con el .add_subplot
#.add_subplot(n°filas, n°columnas, n°grafica)
canalVerdeLuminosidad=figura_canalesLuminosidad.add_subplot(1,3,2)

canalVerdeLuminosidad=np.copy(imagen)

canalVerdeLuminosidad = imagen[:,:,1]

#Ploteamos la imagen
plt.imshow(canalVerdeLuminosidad)

#Seteamos un tamaño mas grande por grafica(lo declaramos una sola vez)
#plt.rcParams["figure.figsize"]=(16,16)

plt.title("Canal Verde")

plt.axis("off")



#-----------------------------IMAGEN CANAL AZUL LUMINOSIDAD--------------------
#Agregamos la imagen a la posicion deseada de la figura con el .add_subplot
#.add_subplot(n°filas, n°columnas, n°grafica)
canalAzulLuminosidad=figura_canalesLuminosidad.add_subplot(1,3,3)

canalAzulLuminosidad=np.copy(imagen)

canalAzulLuminosidad = imagen[:,:,1]

#Ploteamos la imagen
plt.imshow(canalAzulLuminosidad)

#Seteamos un tamaño mas grande por grafica(lo declaramos una sola vez)
#plt.rcParams["figure.figsize"]=(16,16)

plt.title("Canal Azul")

plt.axis("off")
```

```python
'''
Ahora, con una imagen de luminosidad , se aplica el mapa de colores 
(también conocido como tabla de búsqueda, LUT). El predeterminado se llama viridis. 
Hay muchos otros para elegir.
Estamos aplicando el cmap en el canal de luminosidad rojo, se puede aplicar en el canal 
azul y verde tambien.
Link de colormap->https://matplotlib.org/stable/tutorials/colors/colormaps.html
'''
figura_cmap = plt.figure()

#-----------CMAP TIPO HOT-------------

canalRojoLuminosidad_hot = figura_cmap.add_subplot(1,3,1)


canalRojoLuminosidad_hot=imagen[:,:,0]

#Ploteamos la imagen
plt.imshow(canalRojoLuminosidad_hot, cmap="hot")

plt.colorbar(orientation='horizontal')



#-----------CMAP TIPO SPECTRAL-------------

canalRojoLuminosidad_spectral = figura_cmap.add_subplot(1,3,2)


canalRojoLuminosidad_spectral=imagen[:,:,0]

#Ploteamos la imagen
plt.imshow(canalRojoLuminosidad_spectral, cmap="nipy_spectral")

plt.colorbar(orientation='horizontal')

plt.axis("off")


#-----------CMAP TIPO MAGMA-------------

canalRojoLuminosidad_magma = figura_cmap.add_subplot(1,3,3)


canalRojoLuminosidad_magma=imagen[:,:,0]

#Ploteamos la imagen
plt.imshow(canalRojoLuminosidad_magma, cmap="magma")

plt.colorbar(orientation='horizontal')

plt.axis("off")

```
