```python
import matplotlib.pyplot as plt
import numpy as np


imagen=plt.imread('Imagenes/fisica.jpg')

plt.imshow(imagen)

print("Dimensiones de la Imagen: ")
print(imagen.shape)
```

```python
imagenCopiaAzul=np.copy(imagen)

imagenCopiaAzul[:,:,0]=0
imagenCopiaAzul[:,:,1]=0

plt.title("Canal Individual Azul")
plt.imshow(imagenCopiaAzul)
```

```python
imagenCopiaVerde=np.copy(imagen)

imagenCopiaVerde[:,:,0]=0
imagenCopiaVerde[:,:,2]=0

plt.title("Canal Individual Verde")
plt.imshow(imagenCopiaVerde)
```

```python
imagenCopiaRojo=np.copy(imagen)

imagenCopiaRojo[:,:,1]=0
imagenCopiaRojo[:,:,2]=0

plt.title("Canal Individual Rojo")
plt.imshow(imagenCopiaRojo)
```

```python
#Declaramos la figura que contendra las diferentes imagenes de los canales
figura_canalesIndividuales= plt.figure()

#-----------------------------IMAGEN CANAL ROJO INDIVIDUAL--------------------
#Agregamos la imagen a la posicion deseada de la figura con el .add_subplot
#.add_subplot(n°filas, n°columnas, n°grafica)
canalRojoIndividual=figura_canalesIndividuales.add_subplot(1,3,1)

canalRojoIndividual=np.copy(imagen)

canalRojoIndividual[:,:,1]=0
canalRojoIndividual[:,:,2]=0

#Ploteamos la imagen
plt.imshow(canalRojoIndividual)

#Seteamos un tamaño mas grande por grafica(lo declaramos una sola vez)
plt.rcParams["figure.figsize"]=(16,16)

plt.title("Canal Rojo Individual")


#-----------------------------IMAGEN CANAL VERDE INDIVIDUAL--------------------

#Agregamos la imagen a la posicion deseada de la figura con el .add_subplot
#.add_subplot(n°filas, n°columnas, n°grafica)
canalVerdeIndividual=figura_canalesIndividuales.add_subplot(1,3,2)

canalVerdeIndividual=np.copy(imagen)

canalVerdeIndividual[:,:,0]=0
canalVerdeIndividual[:,:,2]=0

#Ploteamos la imagen
plt.imshow(canalVerdeIndividual)

#Seteamos un tamaño mas grande por grafica(lo declaramos una sola vez)
#plt.rcParams["figure.figsize"]=(13,13)

plt.title("Canal Verde Individual")

plt.axis("off")



#-----------------------------IMAGEN CANAL AZUL INDIVIDUAL--------------------

#Agregamos la imagen a la posicion deseada de la figura con el .add_subplot
#.add_subplot(n°filas, n°columnas, n°grafica)
canalAzulIndividual=figura_canalesIndividuales.add_subplot(1,3,3)

canalAzulIndividual=np.copy(imagen)

canalAzulIndividual[:,:,0]=0
canalAzulIndividual[:,:,1]=0

#Ploteamos la imagen
plt.imshow(canalAzulIndividual)

#Seteamos un tamaño mas grande por grafica(lo declaramos una sola vez)
#plt.rcParams["figure.figsize"]=(13,13)

plt.title("Canal Azul Individual")

plt.axis("off")
```
