```python

#newline lectura linea a linea
with open(pathArchivoCSV, newline="") as archivo01:
    
    #delimiter definimos la estructura de nuestro archivo, separado por comas
    #csv.DictReader para diccionarios, csv.reader para listas
    lecturaDeCampos=csv.DictReader(archivo01, delimiter=";")
    
    print('\n------------------Propiedades de Nuestro Archivo CSV-----------------')

    #Obtener Nombre de Campos
    print('\n------------------Nombres de Campos-----------------')

    print('Campos: ',lecturaDeCampos.fieldnames)
    
   

    #Obtener Campos Especificos
    print('\n------------------N° de Empleado y Nombre-----------------')
    
    lecturaRegistros=[]
    
    for row in lecturaDeCampos:
        
        print('N° de Empleado: ',row['N° Empleado'], '| Nombre del Empleado:', row['Nombre'])
        
        #Aprovechamos el for para recorrer todos los registros y almacenarlos en una lista..
        lecturaRegistros.append(row['N° Empleado'])
        lecturaRegistros.append(row['Nombre'])
        lecturaRegistros.append(row['Apellido/s'])
        lecturaRegistros.append(row['Fecha de Nacimiento'])
        lecturaRegistros.append(row['Pais de Nacimiento'])
        lecturaRegistros.append(row['Código de Contrato'])
        lecturaRegistros.append(row['Sueldo'])
      

    #Obtener todos los valores de los Registros
    print('\n------------------Todos los valores de los Registros-----------------')
    
    #Recorremos los registros que almacenamos en la lista anterior..
    for listaRegistros in lecturaRegistros:
        print(listaRegistros)

    
    #Obtener Número de Registros
    print('\n------------------Cantidad de Registros-----------------')

    print('Cantidad de Registros: ',lecturaDeCampos.line_num)

    
    #Obtener Formato
    print('\n------------------Formato del Archivo-----------------')

    print('Formato del Archivo: ',lecturaDeCampos.dialect)
    
    
    #Ordenamietno por Longitud de Elementos(cantidad de caracteres)
    print('\n------------------Ordenamiento por Longitud de Elementos-----------------')

    listaDeCamposOrdenados=sorted(lecturaRegistros, key=len, reverse=False)
    
    print('Ordenamiento por Longitud de Elementos: ',listaDeCamposOrdenados)
    
   #Deberia haber muchos mas metodos para manipulacion
```
