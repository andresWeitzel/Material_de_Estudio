```python
import pandas as pd
import io
import requests


url="https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"

#Creamos un objeto StringIO y decodificamos al formato latino, pero antes usamos el 
#metodo .content para obtener los bytes de la solicitud
solicitud=requests.get(url).content

dataFrame=pd.read_csv(io.StringIO(solicitud.decode('utf-8')))

display(dataFrame)
```

```python
'''
Otra forma de realizar lo mismo pero en vez que el objeto String solicitu un buffer de bytes
lo solicite en texto
'''
from io import StringIO
import pandas as pd
import requests

url='https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv'

solicitud=requests.get(url).text

dataFrame=pd.read_csv(StringIO(solicitud))

display(dataFrame)
```
