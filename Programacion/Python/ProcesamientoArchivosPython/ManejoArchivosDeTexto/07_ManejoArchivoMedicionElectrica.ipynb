```python
'''  
    -Recortamos saltos de lineas, indexamos filas, etc para mejor visualizacion del archivo
    -Creamos una Lista y almacenamos el contenido dentro para la manipulacion
    -SIEMPRE borrar listas previamente si se almacenan en memoria, sino se puede generar una sobrecarga

'''


listaMedicionesElectricas=[]


with open(path, 'r+') as archivo:
    
       
        for i,lineaPorLinea in enumerate(archivo, start=1):
        
            lineaPorLinea=lineaPorLinea.rstrip("\n")

            listaMedicionesElectricas.append(lineaPorLinea + "\n")
            
            print("{}: {}".format(i,lineaPorLinea))

        
```

```python
#Imprimimos el contenido de nuestra lista
print(listaMedicionesElectricas)
```

```python
'''
Vamos a tratar la lista para una mejor visualizacion, al igual que los datos de 
el archivo original, OJO CON EL COPY PASTE, en este for no guardamos nada en la lista
queremos visulizar el contenido de esta
'''


for i,lineaPorLinea in enumerate(listaMedicionesElectricas, start=1):
        
    lineaPorLinea=lineaPorLinea.rstrip("\n")
       
    print("{}: {}".format(i,lineaPorLinea))
```

```python
'''
-Numero de caracteres de la lista Mediciones Electricas
-Notar que cada medicion de cada fila tiene un espacio (\t), por ende podemos dividir cada columna
en listas diferentes y de esta forma tratarlas independientemente
'''
print("Numero de filas de la Lista Medicones Electricas: ",len(listaMedicionesElectricas))
```

```python
'''
Quitamos los encabezados de la lista
Notar que cada fila es un elemento en lista
'''

listaMedicionesElectricas[0:3]=['']
```

```python
for i,lineaPorLinea in enumerate(listaMedicionesElectricas, start=1):
        
    lineaPorLinea=lineaPorLinea.rstrip("\n")
       
    print("{}: {}".format(i,lineaPorLinea))
```

```python
'''
Para colocar cada una las mediciones en una sola columna, vamos a reemplazar los espacios
en blanco por saltos de lineas y a su vez eliminamos los saltos de lineas antes y despues del
reemplazo

'''

listaTension=[]
listaCorriente=[]

for i,lineaPorLinea in enumerate(listaMedicionesElectricas, start=1):
    
    #Hacemos un cambio de caracteres por comodidad
    lineaPorLinea=lineaPorLinea.replace("\n","\t")
    lineaPorLinea=lineaPorLinea.replace("\t"," ")
     
    #Vamos a generar una tupla particionando cuando haya espacios    
    lineaPorLinea=lineaPorLinea.partition(" ")
    
    #Sabamos ahora que todas las mediciones de tension se van a almacenar en los elementos 0 de la tupla
    listaTension.append(lineaPorLinea[0])
    
    listaCorriente.append(lineaPorLinea[2])
    
    
    print("{}: {}".format(i,lineaPorLinea))
```

```python
print(listaTension)
```

```python
'''
Podemos tratar cada lista nuevamente

Lista de Tensiones y Lista de Corrientes

'''

for i,lineaPorLinea in enumerate(listaTension, start=1):
    
    print("{}: {}".format(i,lineaPorLinea))
```

```python
for i,lineaPorLinea in enumerate(listaCorriente, start=1):
    
    print("{}: {}".format(i,lineaPorLinea))
```

```python
'''
Sacamos maximos y minimos de cada lista
'''

print("\n======= Mediciones Analytics =======")

print("\n---Total de Elementos/Mediciones por Tensión/Corriente ---")
print(len(listaTension),"Mediciones de Tension")
print(len(listaCorriente),"Mediciones de Corriente")

print("\n---Consulta de Elementos/Mediciones por Tensión/Corriente ---")
#ATENTI AL FORMATO FUARDADO, LO ULTIMO QUEDO CON COMILLAS SIMPLES Y ESPACIO
print("Medicion de Tension 23v:",'23.00 ' in listaTension)
print("Medicion de Corriente 2.585580e-003A :",'2.585580e-003 ' in listaCorriente)



print("\n---Máxima Tensión/Corriente ---")
print(max(listaTension),"Volts")
print(max(listaCorriente),"Amperes")

'''
SE DEBERIA NORMALIZAR LOS DATOS PARA TRABAJAR CON PORCENTAJES PERO SI SE DIVIDE LO QUE SEA
SOBRE LA CANTIDAD TOTAL DE LOS DATOS OBTENEMOS EL PORCENTAJE DESEADO

'''
print("\n--- Cantidad de Veces que se Repite la Máxima Tensión/Corriente---")
print(listaTension.count(max(listaTension))
    ,"Veces maxima medicion en Tension ("
    ,((listaTension.count(max(listaTension)) * len(listaTension)) / 11002)
    ,"%)"
)

print(listaCorriente.count(max(listaCorriente))
      ,"Veces maxima medicion en Corriente ("
      ,((listaCorriente.count(max(listaCorriente)) * len(listaCorriente)) / 11002)
      ,"%)"
     
     )

print("\n---Minima Tensión/Corriente ---")
print(min(listaTension),"Volts")
print(min(listaCorriente),"Amperes")
```
