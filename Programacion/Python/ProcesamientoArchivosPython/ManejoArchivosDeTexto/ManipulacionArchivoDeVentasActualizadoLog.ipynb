```python
import logging
from datetime import datetime

fechaActual=datetime.now()
fechaActual=fechaActual.strftime('%d/%m/%Y %H:%M:%S')

pathVentas2021="C:/Users/andre/OneDrive/Escritorio/ArchivosPython/Ventas/Ventas2021.txt"
pathVentas2021_Actualizado="C:/Users/andre/OneDrive/Escritorio/ArchivosPython/Ventas/Ventas2021_Actualizado.txt"
pathVentas2021_Actualizado_Log="C:/Users/andre/OneDrive/Escritorio/ArchivosPython/Ventas/Ventas2021_Actualizado_Log.txt"

nuevaLista=[]



#Archivo Original

with open(pathVentas2021,'r+') as archivo:
    
    print('\n==========ARCHIVO MODIFICADO(EXTRACCIÓN DE COMILLAS)=========\n')
   
    for i,lineaPorLinea in enumerate(archivo, start=1):
        
        
        lineaPorLinea=lineaPorLinea.rstrip("\n")
            
        lineaPorLinea=lineaPorLinea.replace('"',"")
        
        nuevaLista.append(lineaPorLinea + "\n")
        
        
        print("{}: {}".format(i,lineaPorLinea))

        
        
        
#Archivo Actualizado y Archivo Log

with open(pathVentas2021_Actualizado_Log,'r+') as archivo02, open(pathVentas2021_Actualizado,'w') as archivo03:
    
    
    
    logging.basicConfig(handlers=[logging.FileHandler(filename=pathVentas2021_Actualizado_Log,
                                                  encoding='utf-8',mode='a+')],
                format="%(asctime)s %(name)s:%(levelname)s:%(message)s",
                datefmt="%d-%m-%Y %H:%M:%S %p",
                level=logging.DEBUG
              )
    
    nuevoCliente01='Miguelina Hamilton, 18/01/2021, $8.600\n'
    nuevoCliente02='Jose Paz, 01/02/2021, $2.000\n'

    
    
    nuevaLista.append('\n------Nuevos Clientes-----\n')

    nuevaLista.append(nuevoCliente01)
    nuevaLista.append(nuevoCliente02)
    nuevaLista.append('\nÚltima Modificación: '+fechaActual)

    archivo03.writelines(nuevaLista)


    logging.info('Se agrega el Cliente: '+nuevoCliente01)

    logging.info('Se agrega el Cliente: '+nuevoCliente02)

    logging.warning('Se Actualizó el Archivo\n')

    logging.info(
                'Se Finaliza la Modificación del Archivo'
                '\n---------------------------------------------\n'
                    )
 


    #Salida Consola
    
    print('\n==========ARCHIVO MODIFICADO(EXTRACCIÓN DE COMILLAS Y NUEVOS CLIENTES)=========\n')

    
    
    for i,lineaPorLinea in enumerate(nuevaLista, start=1):


        lineaPorLinea=lineaPorLinea.rstrip("\n")

        lineaPorLinea=lineaPorLinea.replace('"',"")


        print("{}: {}".format(i,lineaPorLinea))




    
  
        
```
